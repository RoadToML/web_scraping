```
import time
from selenium import webdriver as wb
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup 

options = Options()
options.add_argument('--disable-extensions')
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
driver = wb.Chrome('/usr/lib/chromium-browser/chromedriver', options = options)
```

```
page = driver.get('https://www.nba.com/players')
html_page = driver.page_source
soup = BeautifulSoup(html_page, 'html.parser')

#requests has no method called page_cource, its page.content()
```

```
player_details_list = []

all_players_section = soup.find('section', class_ = 'row nba-player-index__row')
player_details = all_players_section.findAll('a', href=True)
for player in player_details:
  print(player.get_text(separator = ' '))
  player_url= 'https://www.nba.com' + player['href']
  
  #go into new player and gather info
  driver.get(player_url)
  
  player_detail_html = driver.page_source
  player_detail_soup = BeautifulSoup(player_detail_html, 'html.parser')
  
  player_detail_section_tag = player_detail_soup.find('section', class_ = 'nba-player-vitals__bottom menu vertical')
  player_details = player_detail_section_tag.get_text(separator = ' ')
  print(player_details)
```
