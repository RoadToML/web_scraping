```
import time
from selenium import webdriver as wb
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup 

options = Options()
options.add_argument('--disable-extensions')
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
driver = wb.Chrome('/usr/lib/chromium-browser/chromedriver', options = options)
```

```
movie_names = []
movie_rating = []
movie_name_link = []

driver.get('https://www.imdb.com/chart/top?ref_=nv_mv_250')
html_page = driver.page_source

soup = BeautifulSoup(html_page, 'html.parser')
full_table_html = soup.find('table', class_ = 'chart full-width')

for a in full_table_html.findAll('td', class_ ='titleColumn'):
  movie_link = 'https://www.imdb.com' + a.find('a')['href']
  movie_name_link.append(movie_link)
  movie_names.append(a.find('a').get_text())
  
for r in full_table_html.findAll('td', class_='ratingColumn imdbRating'):
  rating_strong = r.find('strong')
  movie_rating.append(rating_strong.get_text())
  
full_list = list(zip(movie_names, movie_rating, movie_name_link)) 
print(full_list)
```

```
small_poster_links = []

for link in movie_name_link:
  driver.get(link)
  soup = BeautifulSoup(driver.page_source, 'html.parser')
  small_poster_div = soup.find('div', class_ = 'poster')
  small_poster_href = 'https://www.imdb.com' + small_poster_div.find('a')['href']
  small_poster_links.append(small_poster_href)
```

```
print(len(small_poster_links))
```

```
large_poster_link_list = []

for poster_link in small_poster_links:
  driver.get(poster_link)
  soup = BeautifulSoup(driver.page_source, 'html.parser')
  
  large_poster_div = soup.findAll('div', class_= 'pswp__zoom-wrap')
  large_poster_src = large_poster_div[1].findAll('img')
  large_poster_link = large_poster_src[1]['src']
  large_poster_link_list.append(large_poster_link)
```

```
print(len(large_poster_link_list))
```
